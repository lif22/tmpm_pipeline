# Plot results from Parameter Study
Use filenames of .pkl files exported by /param_study.ipynb to plot the results of the parameter study.

**Author: Florian Lietz**

```python
import pickle
import pandas as pd
```

```python
# Change "file" variable according to the .pkl filenames to be plotted
# Lemma only
file = "lemma_only3.pkl"
with open(file, "rb") as f:
    lo = pickle.load(f)
# Lemma + Simple
file = "lemma_simple3.pkl"
with open(file, "rb") as f:
    ls = pickle.load(f)

# Lemma + Simple + Bigrams/Trigrams
file = "lemma_simple_bigram_trigram3.pkl"
with open(file, "rb") as f:
    lsb = pickle.load(f)
    
# Lemma + Simple + TF-IDF removal
file = "lemma_simple_tfidf3.pkl"
with open(file, "rb") as f:
    lst = pickle.load(f)
# Lemma + Simple + TF-IDF removal
file = "lemma_simple_bigram_trigram_tfidf3.pkl"
with open(file, "rb") as f:
    lsbt = pickle.load(f)

from_ = 5
to = 30
x = range(from_,to+1)


# Lemma only
y = [x[1] for x in lo]
# Lemma + Simple
y2 = [x[1] for x in ls]
# Lemma + Simple + Bigrams/Trigrams
y3 = [x[1] for x in lsb]
# Lemma + Simple + TF-IDF removal
y4 = [x[1] for x in lst]
# Lemma + Simple + Bigrams/Trigrams + TF-IDF removal
y5 = [x[1] for x in lsbt]


c = pd.DataFrame(y, index=x).T
df_length = len(c)
c.loc[df_length] = y2
df_length = len(c)
c.loc[df_length] = y3
df_length = len(c)
c.loc[df_length] = y4
df_length = len(c)
c.loc[df_length] = y5


p = c.T.plot(lw=2, colormap='jet', marker='.', markersize=10, 
         title='Similarity Scores', grid=True, legend=True, figsize=(20,15))
p.set_xlabel("Number of topics")
p.set_ylabel(r"$S_L$")
p.set_xticks(range(5,30,3))
p.legend(["Lemma Only", "Lemma + Simple", "Lemma + Simple + Bigrams/Trigrams", "Lemma + Simple + TF-IDF Removal","Lemma + Simple + Bigams/Trigrams + TF-IDF Removal"]);

```
